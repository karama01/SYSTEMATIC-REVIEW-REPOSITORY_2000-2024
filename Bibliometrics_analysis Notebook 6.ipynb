WORD CLOUD FROM TITLES

```python
data = pd.read_csv("/content/2_eligible_publications.csv", sep=",", encoding='latin-1')
data
```

```python
# Combine all titles into a single text string
text = " ".join(str(title) for title in data['Title'].dropna())
```

```python
# Preprocess the text: remove non-alphanumeric characters and lower the text
text = re.sub(r'\W+', ' ', text).lower()
```

```python
# Remove stopwords
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in text.split() if word not in stop_words]
```

```python
# Generate a word cloud
wordcloud = WordCloud(
    width=2000,
    height=1000,
    background_color='white',
    stopwords=stop_words,
    max_words=200,
    colormap='viridis'
).generate(" ".join(filtered_words))

# Plot the word cloud
plt.figure(figsize=(16, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
#plt.title("Word Cloud of Titles from Systematic Review", fontsize=16)
plt.show()
```
