```python
mol = papers[papers["Journal"] == "Molecules"] #Molecules
```

```python
mol_year = pd.DataFrame(mol["Year"].value_counts().reset_index())
mol_year.columns = ['year', 'count']
mol_year = mol_year.sort_values(by="year").reset_index(drop=True)
```

```python
mol_year
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
from matplotlib.pyplot import savefig
from matplotlib import rcParams

# ... (Your existing code for data loading and preprocessing)

rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Arial']

fig, ax = plt.subplots(figsize=(14, 8))
ax.xaxis.set_major_locator(MaxNLocator(integer=True))

for axis in ['top', 'bottom', 'left', 'right']:
    ax.spines[axis].set_linewidth(1.5)
ax.xaxis.set_tick_params(width=1.5, labelsize=14)
ax.yaxis.set_tick_params(width=1.5, labelsize=14)

# Set x-axis limits to 2000-2024
ax.set_xlim(2000, 2024)

# Ensure all years from 2000 to 2024 are included in the plot, filling missing years with zero counts
years_to_plot = range(2000, 2025)

# Create complete dataframes for both datasets
per_year_complete = pd.DataFrame({'year': years_to_plot})
per_year_complete = pd.merge(per_year_complete, per_year, on='year', how='left').fillna(0)

mol_year_complete = pd.DataFrame({'year': years_to_plot})
mol_year_complete = pd.merge(mol_year_complete, mol_year, on='year', how='left').fillna(0)


df_count_year_not_top_complete = pd.DataFrame({'Year': years_to_plot})
df_count_year_not_top_complete = pd.merge(df_count_year_not_top_complete, df_count_year_not_top, on='Year', how='left').fillna(0)

pal = sns.color_palette("Set1")
plt.xticks(years_to_plot) # set ticks for all years

plt.plot(per_year_complete['year'], per_year_complete['count'], linewidth=2, label="All publications", ls="dashed")

plt.stackplot(mol_year_complete["year"], df_count_year_not_top_complete["count"], mol_year_complete["count"], labels=['Non-Molecules','Molecules'], colors=pal, alpha=0.4)
plt.legend(loc='upper left', prop={'size': 14})

# Refine x and y ticks for better readability and remove overlaps
plt.xticks(rotation=45, ha='right')
plt.tight_layout() # adjust layout to prevent overlaps

# Display the plot
plt.show()
```
