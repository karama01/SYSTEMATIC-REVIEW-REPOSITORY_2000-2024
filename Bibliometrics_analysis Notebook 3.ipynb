```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = '/content/2_eligible_publications.csv'
data = pd.read_csv(file_path, encoding='ISO-8859-1')

# Define the mapping for Tag1 labels
tag1_labels = {
    "1A": "Antiparasitic",
    "1B": "Antiviral",
    "1C": "Anti-vector",
    "1D": "Antiophidian",
    "1E": "Antifungal",
    "1F": "Antibacterial",
}

# Map the Tag1 codes to their descriptions
data['Tag1_Label'] = data['Tag1'].map(tag1_labels)

# Count the number of papers for each Tag1 label
paper_counts = data['Tag1_Label'].value_counts()

# Plot the results
plt.figure(figsize=(10, 6))
paper_counts.plot(kind='bar', color='skyblue', edgecolor='black')
#plt.title('Paper Counts by Tag1 Labels', fontsize=16)
plt.xlabel('Anti-NTD activities of Natural products', fontsize=14)
plt.ylabel('Number of Papers', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd

# Load the CSV file
file_path = '/content/2_eligible_publications.csv'
data = pd.read_csv(file_path, encoding='ISO-8859-1')

# Define the mapping for Tag1 labels
tag1_labels = {
    "1A": "Antiparasitic",
    "1B": "Antiviral",
    "1C": "Anti-vector",
    "1D": "Antiophidian",
    "1E": "Antifungal",
    "1F": "Antibacterial",
}

# Map the Tag1 codes to their descriptions
data['Tag1_Label'] = data['Tag1'].map(tag1_labels)

# Count the number of papers for each Tag1 label
paper_counts = data['Tag1_Label'].value_counts()

# Create a DataFrame from the paper counts
paper_counts_df = pd.DataFrame({'Tag1_Label': paper_counts.index, 'Number of Papers': paper_counts.values})

# Save the DataFrame to a CSV file
paper_counts_df.to_csv('paper_counts_by_tag.csv', index=False)

print("Paper counts saved to 'paper_counts_by_tag.csv'")
```
