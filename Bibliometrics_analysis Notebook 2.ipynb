```python
# prompt: with table_1_with_abstracts.csv, generate a number of papers per year

import pandas as pd

# Read the CSV file
papers = pd.read_csv("/content/2_eligible_publications.csv", sep=",", encoding='latin-1')

# Create a new DataFrame with the number of papers per year
papers_per_year = papers["Year"].value_counts().reset_index()

# Print the new DataFrame
print(papers_per_year)
```

```python
papers_per_year.to_csv("paper_by_year.csv")
```

```python
df_not_top = papers[(papers["Journal"] != "Molecules")] #all without Molecules
df_not_top
```

```python
df_not_top_by_year = df_not_top.groupby(["Year"])
df_not_top_by_year.head(5)
```

```python
df_count_year_not_top = pd.DataFrame(df_not_top_by_year.describe())["Journal"]["count"]
```

```python
df_count_year_not_top
```

```python
# prompt: df_count_year_not_top to csv

df_count_year_not_top.to_csv("df_count_year_not_top.csv")
```

```python
df_count_year_not_top = df_count_year_not_top.reset_index()
df_count_year_not_top["count"] = df_count_year_not_top["count"].astype("int64")
```

```python
df_count_year_not_top
```

```python
mol = papers[papers["Journal"] == "Molecules"] #Molecules
```

```python
mol_year = pd.DataFrame(mol["Year"].value_counts().reset_index())
mol_year.columns = ['year', 'count']
mol_year = mol_year.sort_values(by="year").reset_index(drop=True)
```

```python
mol_year
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
from matplotlib.pyplot import savefig
from matplotlib import rcParams

# ... (Your existing code) ...

rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Arial']

fig, ax = plt.subplots(figsize=(14,8))
ax.xaxis.set_major_locator(MaxNLocator(integer=True))

# Set the linewidth for the spines
for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(1.5)

# Set tick parameters for clarity
ax.xaxis.set_tick_params(width=1.5, labelsize=14)
ax.yaxis.set_tick_params(width=1.5, labelsize=14)

# Set x-axis limits and ensure 2017 is included
ax.set_xlim(min(per_year['year']), max(per_year['year'])) # Use data min and max
# or explicitly set the limit: ax.set_xlim(2000, 2024) if that's the range you want
plt.xticks(range(int(min(per_year['year'])), int(max(per_year['year'])) + 1)) # ensures all years are displayed


pal = sns.color_palette("Set1")

# Plot the lines
plt.plot(per_year['year'], per_year['count'], linewidth=2, label="All publications", ls="dashed")


# Filter df_count_year_not_top to match the year range in mol_year
# This line ensures that only the years present in both DataFrames are plotted.
df_count_year_not_top = df_count_year_not_top[df_count_year_not_top["Year"].isin(mol_year["year"])]

# Plot the stackplot
plt.stackplot(mol_year["year"], df_count_year_not_top["count"], mol_year["count"], labels=['Non-Molecules','Molecules'], colors=pal, alpha=0.4)
plt.legend(loc='upper left', prop={'size': 14})


plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```
